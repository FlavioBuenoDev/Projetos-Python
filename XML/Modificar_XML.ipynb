# Corrigir Consultas

```python
''' CORRIGIR CONSULTA  ''' 

import xml.etree.ElementTree as ET
import os

# Obter a lista de nomes de arquivo XML no diretório
diretorio = 'C:/Users/flavio.bueno/Projetos_Python/XML'
arquivos_xml = [arquivo for arquivo in os.listdir(diretorio) if arquivo.endswith('.xml')]


# Processar cada arquivo XML
for arquivo in arquivos_xml:
    # Carregar o arquivo XML
    caminho_arquivo = os.path.join(diretorio, arquivo)
    tree = ET.parse(caminho_arquivo)
    root = tree.getroot()

    # Definir o namespace
    namespace = {'ans': 'http://www.ans.gov.br/padroes/tiss/schemas'}    
    
    # Navegar e fazer alterações no XML
    for elemento in root.findall('.//ans:guiaConsulta', namespace):
        numero_carteira = elemento.find('.//ans:numeroCarteira', namespace)

        # Verificar o número de caracteres
        if len(numero_carteira.text) < 6:
            numero_carteira.text = '00' + numero_carteira.text
        
        elif len(numero_carteira.text) < 7:
            numero_carteira.text = '0' + numero_carteira.text  
    
    # Salvar as alterações de volta no arquivo
    tree.write(caminho_arquivo, encoding='UTF-8', xml_declaration=True)
```

```python
import xml.etree.ElementTree as ET
import os

# Definir o namespace
namespace = {
    'ans': 'http://www.ans.gov.br/padroes/tiss/schemas'
    #'xsi': 'http://www.w3.org/2001/XMLSchema-instance'
}

# Obter a lista de nomes de arquivo XML no diretório
diretorio = 'C:/Users/flavio.bueno/Projetos_Python/XML'
arquivos_xml = [arquivo for arquivo in os.listdir(diretorio) if arquivo.endswith('.xml')]

# Processar cada arquivo XML
for arquivo in arquivos_xml:
    # Carregar o arquivo XML
    caminho_arquivo = os.path.join(diretorio, arquivo)
    tree = ET.parse(caminho_arquivo)
    root = tree.getroot()
    
    
# Encontrar todas as guias SP-SADT
guias_sp_sadt = root.findall('.//ns0:guiaSP-SADT', namespace)

# Iterar sobre as guias SP-SADT
for guia_sp_sadt in guias_sp_sadt:
    valor_total_geral_element = guia_sp_sadt.find('.//ns0:valorTotalGeral', namespace)
    procedimentos_executados_element = guia_sp_sadt.find('.//ns0:procedimentosExecutados', namespace)
    valor_proced_geral_element = guia_sp_sadt.find('.//ns0:valorProcedimentos', namespace)
    
    # Verificar se as tags existem
    if valor_total_geral_element is not None and procedimentos_executados_element is not None:
        valor_total = 0.0
        
        # Encontrar todas as tags <ns0:valorTotal> dentro de <ns0:procedimentosExecutados>
        valor_total_elements = procedimentos_executados_element.findall('.//ns0:valorTotal', namespace)
        
        # Somar os valores
        for valor_total_element in valor_total_elements:
            valor = float(valor_total_element.text) if valor_total_element.text else 0.0
            valor_total += valor
        
        # Substituir o valor de <ns0:valorTotalGeral> pelo valor total calculado formatado com duas casas decimais
        valor_total_geral_element.text = '{:.2f}'.format(valor_total)
        valor_proced_geral_element.text = '{:.2f}'.format(valor_total)
        
    # Salvar as alterações de volta no arquivo
tree.write(caminho_arquivo, encoding='UTF-8', xml_declaration=True)
```

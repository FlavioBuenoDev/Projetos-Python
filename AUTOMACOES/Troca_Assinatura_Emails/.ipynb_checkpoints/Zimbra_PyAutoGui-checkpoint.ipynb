# AUTOMAÇÃO PARA ALTERAÇÃO DE ASSINATURA DOS EMAILS

```python
import pandas as pd
import pyautogui
import os
import time

def tab_campo(numero_de_tabs):
    # Defina o número de vezes que deseja pressionar a tecla "tab"

    # Pressione a tecla "tab" usando um loop
    for _ in range(numero_de_tabs):
        press = pyautogui.press("tab")
    time.sleep(0.5)
    return press

# pyautogui.press("f11")

def modificar(email):

    ass_texto = "Teste de assinatura"

    # Campo search
    pyautogui.click(x=1366, y=129)
    pyautogui.write(email)
    time.sleep(1)
    pyautogui.press("enter")
    time.sleep(1)

    # Email
    pyautogui.click(x=348, y=254)
    time.sleep(0.5)
    pyautogui.click(button='right')
    time.sleep(0.5)
    tab_campo(3)
    time.sleep(0.5)
    pyautogui.press('enter')
    time.sleep(3)

    # Preferencias
    pyautogui.click(x=578, y=175)
    time.sleep(1)

    # Assinaturas
    pyautogui.click(x=68, y=376)
    time.sleep(1)

    # Nova Assinatura
    pyautogui.click(x=449, y=556)
    time.sleep(1)
    pyautogui.write("Atual")

    # Formatar como HTML
    pyautogui.click(x=1763, y=309)
    time.sleep(1)
    tab_campo(1)
    pyautogui.press('enter')

    # Entrar codigo fonte
    pyautogui.click(x=625, y=364)
    time.sleep(1)

    # Colar assinatura
    pyautogui.click(x=645, y=301)
    time.sleep(1)
    pyautogui.write(ass_texto)
    pyautogui.click(x=1199, y=889)
    time.sleep(1)

    # Nova mensagem
    pyautogui.click(x=628, y=697)
    time.sleep(1)
    tab_campo(1)
    pyautogui.press('enter')

    # Resposta assinatura
    pyautogui.click(x=765, y=692)
    time.sleep(1)
    tab_campo(1)
    pyautogui.press('enter')

    # Salvar
    pyautogui.click(x=32, y=209)
    time.sleep(1)
    pyautogui.hotkey('ctrl', 'w')
    pyautogui.press('enter')
    time.sleep(1)
    pyautogui.press('f5')
    time.sleep(2)
    pyautogui.press('enter')
    time.sleep(2)
    
time.sleep(3)

def caminho_dados():
    caminho_atual = os.path.abspath('.')
    CAMINHO_DADOS = Path(caminho_atual) / "dados.xlsx"

    tabela= pd.read_excel(CAMINHO_DADOS)
    #display(tabela[['Nome', 'Email']])

    for dados in tabela.index:
        nome = tabela.loc[dados, "Nome"]
        email = tabela.loc[dados, "Email"]
# Chamada da função modificar
        modificar(email)
    
# Chamada da função inicial       
caminho_dados()

print("Terminou.")
```

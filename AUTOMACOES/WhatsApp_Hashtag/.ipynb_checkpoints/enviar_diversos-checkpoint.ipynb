```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

service = Service(ChromeDriverManager().install())
nav = webdriver.Chrome(service=service)
nav.get("https://web.whatsapp.com")
```

```python
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import pyperclip
import pandas as pd

tabela= pd.read_excel("enviar.xlsx")
display(tabela[['Nome', 'Telefone']])

def dig_texto():
    texto = 'texto.txt'
    with open(texto, 'w') as texto:
        texto_dig = input("Digite o texto que deseja enviar: ")
        texto.write(str(texto_dig))

def ler_texto():
    texto = 'texto.txt'
    with open(texto, 'r') as texto:
        return texto.read()

texto_perg = input("Deseja digitar o texto (S ou N): ")

if texto_perg == "S" or texto_perg == "s":
    dig_texto()   

for linha in tabela.index:
    nome = tabela.loc[linha, "Nome"]
    telefone = tabela.loc[linha, "Telefone"]
    
    mensagem = ler_texto()

    link = f"https://web.whatsapp.com/send?phone={telefone}&text={mensagem}"
    nav.get(link)
    
    #espera um elemento que só aparece quando esta carregado
    while len(nav.find_elements('id', "side")) < 1:
        time.sleep(1)
    time.sleep(2)
       
    if len(nav.find_elements('xpath', '//*[@id="app"]/div/span[2]/div/span/div/div/div/div/div/div[1]')) < 1:
        nav.find_element('xpath','//*[@id="main"]/footer/div[1]/div/span[2]/div/div[2]/div[2]/button/span').click()
        time.sleep(5)
    else:
        # adiciona o numero ao arquivo de erros
        arquivo = 'erros.txt'
        with open(arquivo, 'a') as arquivo:
            arquivo.write(str(telefone)+"\n")
        print(f"Número {telefone} invalido")
```

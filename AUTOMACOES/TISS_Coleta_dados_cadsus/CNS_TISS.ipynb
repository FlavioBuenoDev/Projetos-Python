```python
import csv
import os
from pathlib import Path
from openpyxl import Workbook

# Função para coleta de dados no cadsus
def coletarDados(cpf_entrada):
    # Aguarda 2 segundos para que possa mover o mouse para a proxima tela

    time.sleep(2)
    # Preciona tab até o campo Tipo de Documento
    pyautogui.press("f5")

    # Defina o número de vezes que deseja pressionar a tecla "tab"
    numero_de_tabs = 10

    # Pressione a tecla "tab" usando um loop
    for _ in range(numero_de_tabs):
        pyautogui.press("tab")
        time.sleep(0.5)

    # Preenche o campo Tipo de Documento com o CPF
    pyautogui.write("CPF")
    time.sleep(0.5)
    pyautogui.press("enter")
    time.sleep(0.5)

    # Pula para o campo para que seja preenchido com o CPF desejado
    pyautogui.press("tab")
    time.sleep(0.5)
    pyautogui.write(cpf_entrada)
    time.sleep(0.5)
    pyautogui.press("enter")
    time.sleep(2)

    # Cria uma lista vazia para adicionar os dados
    valor = []

    ########### CNS ###################
    # avança até o campo para obter o cns
    pyautogui.click(x=588, y=995)
    # Clica 2x para obter o dado
    pyautogui.click(clicks=2)

    # Copia o dado selecionado
    cns = pyautogui.hotkey('ctrl', 'c')

    # utiliza a biblioteca clipboard para colar o dado copiado acima na variavel cns
    cns = clipboard.paste()
    # adiciona o dado cns na lista
    valor.append(cns)
    time.sleep(0.5)

    ########### NOME ###################
    # avança até o campo para obter o nome
    pyautogui.click(x=747, y=997)
    # Clica 2x para obter o dado
    pyautogui.click(clicks=2)

    # Copia o dado selecionado
    cns = pyautogui.hotkey('ctrl', 'c')

    # utiliza a biblioteca clipboard para colar o dado copiado acima na variavel cns
    nome = clipboard.paste()
    # adiciona o dado cns na lista
    valor.append(nome)
    time.sleep(0.5)

    ########### NASC ###################
    # clica nomeio da barra de rolagem horizontal para exibir o campo de data nasc
    pyautogui.click(x=1263, y=943)
    # clica no valor para selecionar
    pyautogui.click(x=1300, y=996)
    pyautogui.click(clicks=2)
    # copia a data de nasc
    nasc = pyautogui.hotkey('ctrl', 'c')
    # utiliza a biblioteca clipboard para colar o dado copiado acima na variavel nasc
    nasc = clipboard.paste()
    valor.append(nasc)
    time.sleep(0.5)
    
    # Retorna os dados coletados
    return valor[0].strip(), valor[1].strip(), valor[2].strip()
    
    
    
# Cria um novo livro (arquivo Excel)
wb = Workbook()

# Adiciona uma nova planilha
planilha = wb.create_sheet("Beneficiários")

# Acessa a planilha ativa (por padrão, a primeira planilha criada)
planilha_ativa = wb.active

caminho_atual = os.path.abspath('.')
CAMINHO_CSV = Path(caminho_atual) / "BENEFICIARIO_CNS.csv"

with open(CAMINHO_CSV, 'r') as arquivo:
    # Usando DictReader diretamente com o ponto e vírgula como delimitador
    leitor_csv = csv.DictReader(arquivo, delimiter=';')

    # Iterando sobre as linhas do arquivo CSV como dicionários
    for indice, linha in enumerate(leitor_csv, start=2):
        # Adiciona os dados às células da planilha
        planilha[f'A{indice}'] = linha['MATRICULA']
        planilha[f'B{indice}'] = linha['NOME']
        planilha[f'C{indice}'] = linha['CPF']

        # Chama a função coletarDados e obtém os valores
        cns, nome, nasc = coletarDados(linha['CPF'])

        # Adiciona os valores obtidos na planilha
        planilha[f'D{indice}'] = cns
        planilha[f'E{indice}'] = nome
        planilha[f'F{indice}'] = nasc

# Salva o livro em um arquivo
wb.save("BENEFICIARIOS_CNS_SAIDA.xlsx")

print("Dados adicionados com sucesso.")
```
